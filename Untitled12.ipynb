```python
# ==========================================================
# AI-POWERED SALES FORECASTING - TASK 1 (SINGLE FILE)
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# -------------------------------
# STEP 1: CREATE SAMPLE DATASET
# -------------------------------

np.random.seed(42)

dates = pd.date_range(start="2019-01-01", end="2023-12-31", freq="D")
sales = np.random.randint(200, 500, size=len(dates)) + np.linspace(0, 300, len(dates))

data = pd.DataFrame({
    "date": dates,
    "sales": sales.astype(int)
})

print("Dataset Created Successfully!")
print(data.head())

# -------------------------------
# STEP 2: EXPLORATORY DATA ANALYSIS
# -------------------------------

print("\nDataset Info:")
print(data.info())

print("\nStatistical Summary:")
print(data.describe())

plt.figure(figsize=(10, 4))
plt.plot(data["date"], data["sales"])
plt.title("Historical Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()

# -------------------------------
# STEP 3: SIMPLE FORECASTING MODEL
# (Moving Average – No extra installs)
# -------------------------------

window_size = 30
data["moving_avg"] = data["sales"].rolling(window=window_size).mean()

# -------------------------------
# STEP 4: FUTURE SALES PREDICTION
# -------------------------------

future_days = 30
last_date = data["date"].iloc[-1]

future_dates = [last_date + timedelta(days=i) for i in range(1, future_days + 1)]
future_sales = [data["sales"].iloc[-window_size:].mean()] * future_days

forecast = pd.DataFrame({
    "date": future_dates,
    "predicted_sales": future_sales
})

print("\nFuture Sales Forecast:")
print(forecast.head())

# -------------------------------
# STEP 5: VISUALIZATION
# -------------------------------

plt.figure(figsize=(10, 4))
plt.plot(data["date"], data["sales"], label="Actual Sales")
plt.plot(data["date"], data["moving_avg"], label="Moving Average", linestyle="--")
plt.plot(forecast["date"], forecast["predicted_sales"], label="Forecast", color="red")
plt.legend()
plt.title("Sales Forecasting Dashboard")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()

# -------------------------------
# STEP 6: SAVE OUTPUT (OPTIONAL)
# -------------------------------

data.to_csv("historical_sales.csv", index=False)
forecast.to_csv("future_sales_forecast.csv", index=False)

print("\nFiles Saved:")
print("✔ historical_sales.csv")
print("✔ future_sales_forecast.csv")

# -------------------------------
# END OF FILE
# -------------------------------
```
